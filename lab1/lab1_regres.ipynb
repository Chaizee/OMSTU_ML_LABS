```python
import pandas as pd

data_whitewine = pd.read_csv('../csv/winequality-white.csv', sep = ';')
data_whitewine
data_whitewine['red'] = 0
```

```python
data_whitewine
```

```python
data_redwine = pd.read_csv('../csv/winequality-red.csv', sep = ';')
data_redwine
data_redwine['red'] = 1
```

```python
data_redwine
```

```python
data = pd.concat([data_redwine, data_whitewine], axis = 0)
data
```

```python
data = data[['red'] + [col for col in data.columns if col != 'red']]
data
```

```python
data.isnull().sum()
```

```python
import seaborn as sns

sns.heatmap(data.corr())
```

```python
sns.countplot(data=data, x='quality', hue='red')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.histplot(data=data, x='alcohol', hue='red', kde=True, ax=axes[0,0])
sns.histplot(data=data, x='residual sugar', hue='red', kde=True, ax=axes[0,1])
sns.histplot(data=data, x='volatile acidity', hue='red', kde=True, ax=axes[1,0])
sns.histplot(data=data, x='sulphates', hue='red', kde=True, ax=axes[1,1])

plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='quality', y='alcohol', hue='red')
plt.title('Alcohol vs Quality (по типу вина)')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='quality', y='residual sugar', hue = 'red')
plt.title('Alcohol vs Quality')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='quality', y='volatile acidity', hue = 'red')
plt.title('Alcohol vs Volatile acidity')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='quality', y='sulphates', hue = 'red')
plt.title('Alcohol vs Sulphates')
```

```python
data.to_csv("../csv/winequality_finished.csv", index= False, sep=';')
```
